```python
import tensorflow as tf
```

```python
#chapter2-1
g1 = tf.Graph()
with g1.as_default():
    a = tf.constant(1)
    b = tf.constant(3)
    c = tf.add(a,b)
```

```python
with tf.Session(graph=g1) as sess1:
    print(sess1.run(c))
```

```python
with tf.Session(graph=g1) as sess2:
    print(sess2.run(c))
```

```python
#chapter2-2
x = tf.constant(4)
y = tf.constant(5)
x1 = tf.constant(4.2)
y1 = tf.constant(5.1)
result_add = tf.add(x,y)
result_subtract = tf.subtract(x,y)
result_multiply = tf.multiply(x,y)
result_divide = tf.divide(x,y)
result_mod = tf.mod(x1,y1)
result_pow = tf.pow(x,y)
result_square = tf.square(x)
result_sqrt = tf.sqrt(x1)

with tf.Session() as sess:
    print(sess.run(result_add))
    print(sess.run(result_subtract))
    print(sess.run(result_multiply))
    print(sess.run(result_divide))
    print(sess.run(result_mod))
    print(sess.run(result_pow))
    print(sess.run(result_square))
    print(sess.run(result_sqrt))
```

```python
m = tf.constant([[1,2],[4,5]])
n = tf.constant([[5,6],[7,1]])
result_pow = tf.pow(m,n)
with tf.Session() as sess1:
    print(sess1.run(result_pow))
```

```python
#chapter3
g = tf.Graph()
with g.as_default():
    tf.set_random_seed(1)
    img = tf.random_uniform([10,28,28,3])
```

```python
#chapter3-1
select_2 = img[1,:,:,:]
with tf.Session(graph=g) as sess:
    print(sess.run(select_2))
```

```python
#chapter3-2
slice_select_2 = tf.slice(img,[1,0,0,0],[1,28,28,3])
with tf.Session(graph=g) as sess:
    print(sess.run(slice_select_2))
```

```python
#chapter3-3
    #rank is different
#chapter3-4
for i in range(1,5):
    select_1357 = tf.slice(img,[2*i-1,0,0,0],[1,28,28,3])
    with tf.Session(graph=g) as sess:
        print(sess.run(select_1357))
```

```python
#chapter3-5
select_67 = tf.slice(img,[5,7,7,0],[2,14,14,3])
with tf.Session(graph=g) as sess:
    print(sess.run(select_67))
```

```python
#chapter3-6
split_1 = tf.slice(img,[0,0,0,0],[10,28,28,1])
split_2 = tf.slice(img,[0,0,0,1],[10,28,28,1])
split_3 = tf.slice(img,[0,0,0,2],[10,28,28,1])
with tf.Session(graph=g) as sess:
    print(sess.run(split_1))
    print(sess.run(split_2))
    print(sess.run(split_3))
```

```python
#chapter3-7
a = tf.shape(tf.shape(img))
b = tf.rank(tf.shape(img))
c = tf.shape(img)
d = tf.rank(img)
with tf.Session(graph=g) as sess:
    print(sess.run(a))
    print(sess.run(b))
    print(sess.run(c))
    print(sess.run(d))
```
